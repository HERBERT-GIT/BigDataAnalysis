```python
import pandas as pd # importing pandas library
import numpy as np # importing numpy library
```

```python
df= pd.read_csv('/content/sample_data/fashion dataset.csv')
df
```

```python
df
```

```python
df1= pd.read_csv('/content/sample_data/fashion brand details.csv')
df1
```

```python
df.info()
```

```python
df.drop(df.columns[[7,8]], axis=1, inplace=True)
df
```

```python
df1.rename(columns = {'brand_name':'brand'}, inplace = True)
df1
```

```python
df['brand'] = df['brand'].str.lower()
df1['brand'] = df1['brand'].str.lower()
```

```python
newdf= pd.merge(df,df1, left_on='brand', right_on='brand')
newdf
```

```python
newdf.info()
```

```python
newdf_mean =newdf
newdf_mean['ratingCount']=newdf_mean['ratingCount'].fillna(newdf_mean['ratingCount'].mean())
newdf_mean['avg_rating']=newdf_mean['avg_rating'].fillna(newdf_mean['avg_rating'].mean())
newdf_mean.info()
```

```python
newdf_mean
```

```python
newdf_zero =newdf
newdf_zero['ratingCount']=newdf_zero['ratingCount'].fillna(0)
newdf_zero['avg_rating']=newdf_zero['avg_rating'].fillna(0)
newdf_zero.info()
```

```python
newdf_zero
```

```python
newdf_zero=newdf_zero.drop_duplicates()
newdf_zero
```

```python
newdf_zero.info()
```

```python
newdf_zero = newdf_zero.dropna(axis=0)
newdf_zero.info()
```

```python
from google.colab import files

newdf_zero.to_csv('df.csv')
files.download('df.csv')
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
```

```python
newdf_zero
```

```python
km=KMeans(n_clusters=3, random_state=0)
```

```python
newdf_zero=newdf_zero[['brand_id', 'avg_rating']]
```

```python
predict_y=km.fit_predict(newdf_zero)
```

```python
predict_y
```

```python
km.cluster_centers_
```

```python
newdf_zero['cluster']=predict_y
newdf_zero
```

```python
df_c0= newdf_zero[newdf_zero.cluster==0]
df_c1=newdf_zero[newdf_zero.cluster==1]
df_c2=newdf_zero[newdf_zero.cluster==2]

plt.scatter(df_c0['brand_id'], df_c0['avg_rating'], label='Cluster 0', color='green')
plt.scatter(df_c1['brand_id'], df_c1['avg_rating'], label='Cluster 1',color='blue')
plt.scatter(df_c2['brand_id'], df_c2['avg_rating'], label='Cluster 2',color='purple')

plt.xlabel('brand_id')
plt.ylabel('avg_rating')
plt.legend()
```

```python
scaler=MinMaxScaler()
scaler.fit(newdf_zero[['avg_rating']])
newdf_zero[['avg_rating']]= scaler.transform(newdf_zero[['avg_rating']])
newdf_zero
```

```python
scaler.fit(newdf_zero[['brand_id']])
newdf_zero[['brand_id']]= scaler.transform(newdf_zero[['brand_id']])
newdf_zero
```

```python
km= KMeans(n_clusters=3, random_state=0)
```

```python
df2= newdf_zero[['avg_rating','brand_id']]
df2
```

```python
predict_y = km.fit_predict(df2)
predict_y
```

```python
df2['New_clusters']=predict_y
df2
```

```python
df_c02=df2[df2.New_clusters==0]
df_c12=df2[df2.New_clusters==1]
df_c22=df2[df2.New_clusters==2]


plt.scatter(df_c02['brand_id'], df_c02['avg_rating'], label='Cluster 0', color='green')
plt.scatter(df_c12['brand_id'], df_c12['avg_rating'], label='Cluster 1', color='blue')
plt.scatter(df_c22['brand_id'], df_c22['avg_rating'], label='Cluster 2', color='purple')


plt.ylabel('avg_rating')
plt.xlabel('brand_id')
plt.legend()
```

```python
km.cluster_centers_
```

```python
df_c02=df2[df2.New_clusters==0]
df_c12=df2[df2.New_clusters==1]
df_c22=df2[df2.New_clusters==2]


plt.scatter(df_c02['brand_id'], df_c02['avg_rating'], label='Cluster 0', color='green')
plt.scatter(df_c12['brand_id'], df_c12['avg_rating'], label='Cluster 1', color='blue')
plt.scatter(df_c22['brand_id'], df_c22['avg_rating'], label='Cluster 2', color='purple')
plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:,1], color='black',  marker='X', label ='centroid')

plt.ylabel('avg_rating')
plt.xlabel('brand_id')
plt.legend()
```

```python
#Elbow method
#Calculate Sum of Square Error(SSE)

k_range=range(1,10)
sse=[]
for k in k_range:
    km = KMeans(n_clusters=k)
    km.fit(newdf_zero[['brand_id', 'avg_rating']])
    sse.append(km.inertia_) 

sse
```

```python
plt.xlabel('K')
plt.ylabel('SSE')
plt.plot(k_range,sse)
```
